```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
group_boudaries =[70, 150, 250, 400, 600, 900, 1400, 2000]
```

```python
plt.plot(group_boudaries)
plt.show()
```

```python
density_x = [0] + group_boudaries
```

```python
n = len(density_x)
density_y = [1-k/n for k in range(n)]
```

```python
x = np.array(density_x)
y = np.array(density_y)
```

```python
a, b = np.polyfit(x, np.log(y), 1)
```

```python
y_reg = np.exp(a*x + b)
```

```python
plt.plot(x, y)
plt.plot(x, y_reg, '--')
plt.legend(['Densité perçue', f'Régression exponentielle exp({a:.5f}x {b:.5f})'])
plt.title("Densité de temps de survie")
plt.xlabel("Jours")
plt.ylabel("Densité")
plt.savefig(f'density_survival.png')
plt.show()
```

```python
a2, b2 = np.polyfit(x, 1/y, 1)
```

```python
y_reg2 = 1/(a2*x + b2)
```

```python
plt.plot(x, y)
plt.plot(x, y_reg2, '--')
plt.show()
```
