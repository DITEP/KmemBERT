```python
model = CamembertForSequenceClassification.from_pretrained("camembert-base", num_labels=2, output_attentions=True)
tokenizer = CamembertTokenizerFast.from_pretrained("camembert-base")
```

## Function to visualise attention

```python
def show_attention(sentence):
    inputs = tokenizer.encode(sentence, return_tensors='pt')
    outputs = model(inputs)
    attention = outputs[-1]  
    tokens = tokenizer.convert_ids_to_tokens(inputs[0])
    head_view(attention, tokens) 
```

#### Change the sentence variable to visualise the attention mechanism on it. 

```python
sentence  = "La tumeur se développe"
show_attention(sentence)
```
